```python
def convertUnits(d, oldUnit, newUnit, k) :
    for ingredient in d :
        if d[ingredient]["unit"] == oldUnit :
            d[ingredient]["unit"] = newUnit
            d[ingredient]["amt"] *= k
    return d
            
def getIngredients(soup) :
    newSoup = soup.find_all('div', class_='card-body')[0].find_all('div', class_='row')
    d = dict()
    for thing in newSoup :
        l = thing.contents
        if len(l) != 3 :
            #print(f"len(l) = {len(l)} != 2 -> continuing")
            continue
        name = l[1].text
        d[name] = {"amt": float(l[0].contents[0]), "unit": l[0].contents[1].text}
    return {"ingredients": convertUnits(d, "ml", "cL", .1)}

def getGlassAndMethod(soup) :
    newSoup = soup.find_all('div', class_='card-body mt-3')[1]
    glass = newSoup.find_all('div', class_='col-6 text-center')[1].find_all('span')[0].text
    method = newSoup.find_all('div', class_='col-12')[0].text
    return {"glass": glass, "method": method}

def getInstructionsAndNotes(soup) :
    newSoup = soup.find_all("div", class_="card-body mt-3 text-left")[0]
    i = 1
    instructions = dict()
    for step in newSoup.find_all("li") :
        instructions[i] = step.text
        i += 1
    notes = newSoup.find_all("p")[0].text
    return {"instructions": instructions, "notes":notes}
```

```python
def getRecipe(cocktailName) :
    url = "http://kuletos.com.au/cocktails/How-To-Make-A/" + cocktailName.replace(" ", "-")
    rq = requests.get(url)
    if rq.status_code != 200 :
        print(f"ERROR : GET '{url}' -> error #{rq.status_code}")
        return 
    content = BeautifulSoup(rq.content, 'html.parser')
    wholeDict = dict()
    wholeDict.update(getIngredients(content))
    wholeDict.update(getGlassAndMethod(content))
    wholeDict.update(getInstructionsAndNotes(content))
    return wholeDict
    
```

```python
itemAmount = 5731
itemsPerPage = 100
url = f"https://www.diffordsguide.com/cocktails/search?include%5Bdg%5D=1&include%5Bdiscerning_drinker%5D=1&include%5Bcocktails_15%5D=1&include%5Ball%5D=1&sort=rating&limit={itemsPerPage}&offset=0"
rq = requests.get(url)
soup = BeautifulSoup(rq.content, 'html.parser')
```

```python
def getIngredients(cells) :
    ingredients = dict()
    for t in cells[3].find("table").find_all("tr") :
        if t.find_all("td")[0].text.strip(' \n\t').startswith("Loading...") :
            break
        qty = t.find_all("td")[0].text.strip(' \n\t').split(" ")[0]
        unit = t.find_all("td")[0].text.strip(' \n\t').split(" ")[1]
        what = t.find_all("td")[1].text.strip(' \n\t')
        ingredients[what] = {"quantity": qty, "unit": unit}
    return ingredients
    
def getRatings(soup) :
    totalValueCheck = 5
    
    ratings = soup.find("div", class_="grid-x grid-margin-x align-center-middle justify-left-small")

    if ratings == None :
        return {"publisher": {"name": None, "rating": None}, "users": {"n": None, "rating": None}}

    if len(ratings.contents) > 1 :
        diffordsRating = ratings.contents[1]

        name = diffordsRating.find("div", class_="cell shrink alt-font text-small pad-right-half no-wrap").text.strip(' \n\t')

        total = 0
        note = 0
        for star in diffordsRating.find_all("span", class_=re.compile('^svg-icon--star.*')) :
            total += 1
            if "svg-icon--star" in star["class"] :
                note += 1
            elif "svg-icon--star-half" in star["class"] :
                note += .5
            elif "svg-icon--star-empty" in star["class"] :
                note += 0 
            else :
                print(f"Error no 'good' class in {star['class']}")

    if total == totalValueCheck :
        dgNote = note/total 
    else :
        dgNote = None

    if len(ratings.contents) <= 5 :
        return {"publisher": {"name": "difford", "rating": dgNote}, "users": {"n": None, "rating": None}}

    usersRating = ratings.contents[5]
    name = usersRating.find("div", class_="cell shrink alt-font text-small always-first-small no-wrap").text.strip(' \n\t')

    nbOfRatings = name.split("(")[-1].split(" ")[0]
    name = name.split(" (")[0]

    total = 0
    note = 0
    for star in usersRating.find_all("span", class_=re.compile('^svg-icon--star.*')) :
        total += 1
        if "svg-icon--star" in star["class"] :
            note += 1
        elif "svg-icon--star-half" in star["class"] :
            note += .5
        elif "svg-icon--star-empty" in star["class"] :
            note += 0 
        else :
            print(f"Error no 'good' class in {star['class']}")

    if total == totalValueCheck :
        usersNote = note/total 
    else :
        usersNote = None

    return {"publisher": {"name": "difford", "rating": dgNote}, "users": {"n": nbOfRatings, "rating": usersNote}}



def getRecipeFromDifford(url, driver) :
    driver.get("https://www.diffordsguide.com/cocktails/recipe/1728/st-kitts-non-alcoholic")
    rq = requests.post(url, data = {"display":"ml"})
    soup = BeautifulSoup(rq.content, 'html.parser') 
    cocktail = dict()
    cocktail["name"] = soup.find("h1", class_="strip__heading").text
    cocktail["picture"] = None
    for thing in soup.find_all("img", alt=True) :
        if thing["alt"] == "Cocktail image" or thing["alt"] == "Product image" :
            cocktail["picture"] = thing["src"]
            break
    cells = soup.find("article", class_="cell long-form long-form--small long-form--inline-paragraph pad-bottom").find_all("div", class_="cell")
    cocktail["glass"] = cells[0].find("a").text
    cocktail["garnish"] = cells[1].find("p").text
    cocktail["instructions"] = cells[2].find("p").text
    cocktail["ingredients"] = getIngredients(cells)
    cocktail["rating"] = getRatings(soup)
    return cocktail
```

```python
import os.path

def getPictures(df, source) :
    
    totalItems = len(df.index)
    itemsPerPage = 100
    cocktails = []
    itemCount = 0
    startTime = time.time()
    
    pictures = []
    
    picturesFolder = f"cocktails/{source}/pictures/"
    for ind, row in df.iterrows() :
        url = row["picture"]
        fileName = "".join(x for x in row['name'].replace(' ', '_') if (x.isalnum() or x == "_"))
        
        if (url != None) :
            
            picturePath = f"{picturesFolder}{fileName}.{url.split('.')[-1]}"

            if os.path.isfile(picturePath) :

                print(f"Skipping '{picturePath}'... (already there)")
                pictures.append(picturePath)

                totalItems -= 1

                if itemCount > 0 and itemCount % 100 == 0 :

                    display.clear_output(wait=True)
                    totalItemsString = "{:,}".format(totalItems).replace(",", " ")
                    countItemsString = "{:,}".format(itemCount).replace(",", " ")
                    percentageString = "{:.4}%".format(100.0 * itemCount / totalItems)
                    timeElapsed = time.time() - startTime
                    eta = int((totalItems - itemCount) * timeElapsed / itemCount)
                    string = (f"{countItemsString} / {totalItemsString} done ({percentageString})" + 
                        f" > {hmsString(timeElapsed)} : {hmsString(eta)} ({hmsString(timeElapsed + eta)})")
                    print(string)

            else :

                print(f"Creating '{picturePath}'...")

                response = requests.get(url)
                file = open(picturePath, "wb")
                file.write(response.content)
                file.close()
                pictures.append(picturePath)

                itemCount += 1

                display.clear_output(wait=True)
                totalItemsString = "{:,}".format(totalItems).replace(",", " ")
                countItemsString = "{:,}".format(itemCount).replace(",", " ")
                percentageString = "{:.4}%".format(100.0 * itemCount / totalItems)
                timeElapsed = time.time() - startTime
                eta = int((totalItems - itemCount) * timeElapsed / itemCount)
                string = (f"{countItemsString} / {totalItemsString} done ({percentageString})" + 
                    f" > {hmsString(timeElapsed)} : {hmsString(eta)} ({hmsString(timeElapsed + eta)})")
                print(string)
                
        else : 
            
            print(f"Skipping '{picturePath}'... (no URL)")
            pictures.append(None)

            totalItems -= 1

            if itemCount > 0 and itemCount % 100 == 0 :

                display.clear_output(wait=True)
                totalItemsString = "{:,}".format(totalItems).replace(",", " ")
                countItemsString = "{:,}".format(itemCount).replace(",", " ")
                percentageString = "{:.4}%".format(100.0 * itemCount / totalItems)
                timeElapsed = time.time() - startTime
                eta = int((totalItems - itemCount) * timeElapsed / itemCount)
                string = (f"{countItemsString} / {totalItemsString} done ({percentageString})" + 
                    f" > {hmsString(timeElapsed)} : {hmsString(eta)} ({hmsString(timeElapsed + eta)})")
                print(string)
    
    return pictures
```

```python
def isARecipe(soup) :
    words = ["ORIGIN", "GLASSWARE", "GARNISH", "SOURCE", "NOTE"]
    n = 0
    threshold = 2
    
    for word in words :
        if (f"\n{word}" in soup.text) :
            n += 1
    return n >= threshold

def mayBeARecipe(soup) :
    words = ["ORIGIN", "GLASSWARE", "GARNISH", "SOURCE", "NOTE"]
    n = 0
    for word in words :
        if (f"\n{word}" in soup.text) :
            n += 1
    return n

def cutRedoundantNewLines(text) :
    old = soup.text
    new = old.replace("\n\n", "\n")
    while old != new :
        #print(f"old = '{old}', new = '{new}'")
        old = new
        new = old.replace("\n ", "\n")
        new = old.replace("\n\n", "\n")
    return new
```

```python
def getRecipeFromSC(soup) :
    cocktail = dict()
    soup = soup.find("body")
    cocktail["ingredients"] = []
    for candidate in soup.find_all("p") :
        for htmlClass in candidate.attrs["class"] :
            if (htmlClass == "rt") :
                cocktail["name"] = candidate.text
                break
            elif ("il" in htmlClass) :
                cocktail["ingredients"].append(candidate.text)
                break
            elif (("ry" in htmlClass) or ("garnish" in htmlClass) or ("rx" in htmlClass)) :
                span = candidate.find("span")
                if span == None :
                    cocktail["instructions"] = candidate.text
                elif span.text == "ORIGIN" :
                    cocktail["origin"] = candidate.text.replace("ORIGIN ", "")
                elif span.text == "GLASSWARE" :
                    cocktail["glass"] = candidate.text.replace("GLASSWARE ", "")
                elif span.text == "GARNISH" :
                    cocktail["garnish"] = candidate.text.replace("GARNISH ", "")
                elif span.text == "SOURCE" :
                    cocktail["source"] = candidate.text.replace("SOURCE ", "")
                elif span.text == "NOTE" :
                    cocktail["note"] = candidate.text.replace("NOTE ", "")
    return cocktail
            
def useless(cocktail, soup) :
    cocktail["name"] = soup.find("h1", class_="strip__heading").text
    cocktail["picture"] = None
    for thing in soup.find_all("img", alt=True) :
        if thing["alt"] == "Cocktail image" or thing["alt"] == "Product image" :
            cocktail["picture"] = thing["src"]
            break
    cells = soup.find("article", class_="cell long-form long-form--small long-form--inline-paragraph pad-bottom").find_all("div", class_="cell")
    cocktail["glass"] = cells[0].find("a").text
    cocktail["garnish"] = cells[1].find("p").text
    cocktail["instructions"] = cells[2].find("p").text
```
